```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices2.csv")
df.head()
```

```python
#to fix the null value we find the median of all bedroom values and put floor that means no decimal points
import math
median_bedrooms=math.floor(df.bedrooms.median())
print(median_bedrooms)
```

```python
df.bedrooms=df.bedrooms.fillna(median_bedrooms)#now null (na) is gone
df.head()
```

```python
#our dataset is now perfect we can just train it now , price is a dependent variable on 3 independent variables atrea , bedrooms,age
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
print(reg.coef_)
print(reg.intercept_)
```

```python
#we can now predict the values
reg.predict([[3000,4,35]])

```
